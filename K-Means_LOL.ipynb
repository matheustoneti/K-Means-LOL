```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import cluster
```

```python
dados = pd.read_csv('high_diamond_ranked_10min.csv')
dados.head()
```

```python
dados_cluster = dados[['blueWins', 'blueFirstBlood', 'blueKills', 'blueDeaths', 'blueAssists']]
sns.pairplot(dados_cluster, hue='blueWins')
```

```python
fig, axe = plt.subplots (figsize = (8,5))
axe = sns.scatterplot (data=dados_cluster, x='blueAssists', y='blueKills', hue='blueWins')
```

```python
kmeans = cluster.KMeans (2)
clusters = kmeans.fit_predict(dados_cluster)
pd.value_counts(clusters)
```

```python
x_center, y_center = kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1]
fig, (axe1) = plt.subplots(1, figsize=(14,5))
sns.scatterplot (data = dados_cluster, x='blueAssists', y='blueKills', hue= clusters, ax=axe1)

```
